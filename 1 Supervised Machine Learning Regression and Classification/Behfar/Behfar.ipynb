```python
def model1(metrage: int) -> int:
    price = metrage * 1000
    return price

def model2(metrage, price):
    if price > 400000:
        result = "Iran"
    elif price > 300000:
        result = "USA"
    else:
        result = "India"
    return result

# Input
input_metrage = 500

# Prediction
predicted_price = model1(input_metrage)

predicted_country = model2(input_metrage, predicted_price)
predicted_country = model2(input_metrage, input_metrage * 1000)
```
