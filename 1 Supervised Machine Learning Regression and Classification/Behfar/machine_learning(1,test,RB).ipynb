```python
import numpy as np
import matplotlib.pyplot as plt
#data----------
data_x = np.array([1,2,3,4,5])
data_y = np.array([2,4,6,8,9])
#--------------


#------set------
w=2
b=0
#---------------

#----formulas---
def test(x,y):
    Y= w*x+b
    e=(y-Y)**2
    return Y,e
Y,e = test(data_x,data_y)
#---------------

#---printdata---
print("Y:",Y,"\ne:",e)
print("j(w,b):",sum(e))

#-----plot-----
plt.scatter(data_x, data_y)
plt.show()
```
